### 1.1.1. Measuring image smoothness

```python
# Load original image as int datatype, due to overflow.
FibresImage = skimage.io.imread("fibres_xcth.png").astype(np.int);

# Implement variation function:
def TotalVariance(Im):
    [w,h] = Im.shape
    Variation = 0
    for x in range(1,w):
        for y in range(1,h):
            # For every cell, take the difference of the cells in previous row and column.
            Variation += abs(Im[x,y] - Im[x-1,y]) + abs(Im[x,y] - Im[x,y-1])
    return Variation

# Show image and total variation.
plt.figure(num=None, figsize=(2, 2), dpi=200, facecolor='w', edgecolor='k')
plt.imshow(FibresImage, cmap = 'gray');
plt.title("Original Image: \n Total Variation =" + "%.3g" % TotalVariance(FibresImage));
plt.savefig("Fibres1.jpg")

```

```python
# Gaussian smoothing: make use of the Gaussian filter from skimage.
FibresImageBlurry = skimage.filters.gaussian(FibresImage, sigma=1, output=None, preserve_range=True)

plt.figure(num=1, figsize=(10, 4), dpi=200, facecolor='w', edgecolor='k')
plt.subplot(1, 3, 1)
plt.imshow(FibresImage, cmap = 'gray');
plt.title("Original Image: \n Total Variation =" + "%.3g" % TotalVariance(FibresImage));

plt.subplot(1, 3, 2)
plt.imshow(FibresImageBlurry, cmap = 'gray');
plt.title("Blurred Image: $\sigma$ = 1 \n Total Variation =" + "%.3g" % TotalVariance(FibresImageBlurry));

plt.subplot(1, 3, 3)
FibresImageBlurrier = skimage.filters.gaussian(FibresImage, sigma=2, output=None, preserve_range=True)
plt.imshow(FibresImageBlurrier, cmap = 'gray');
plt.title("Blurred Image: $\sigma$ = 8 \n Total Variation =" + "%.3g" % TotalVariance(FibresImageBlurrier));
plt.savefig("Fibres1.jpg")



# The decrease in variation is quite clear.
```

### 1.1.2. Computing boundary length



```python
FC1 = skimage.io.imread("fuel_cell_1.tif");
FC2 = skimage.io.imread("fuel_cell_2.tif");
FC3 = skimage.io.imread("fuel_cell_3.tif");

# Define function for the boundary length
def BoundaryLength(Im):
    [w,h] = Im.shape
    Length = 0
    for x in range(1,w):
        for y in range(1,h):
            if (Im[x,y] != Im[x-1,y]):
                Length += 1
    return Length

# As we zoom in on the original image, we can see that the boundary length decreases as expected.
print(BoundaryLength(FC1))
print(BoundaryLength(FC2))
print(BoundaryLength(FC3))
```

### 1.1.3. Curve Smoothing



```python
# Load dino.txt, append first point for a closed curve and load noise. The original dino will always be shown in green.
Dino = np.loadtxt(fname = "dino.txt")
Dino = np.append(Dino, [Dino[0,]], axis=0)
DiNoise = np.loadtxt(fname = "dino_noisy.txt")
DiNoise = np.append(DiNoise, [DiNoise[0,]], axis=0)
#DiNoise = Dino + (0.5 - np.random.rand(len(Dino),2))

plt.figure(num=2, figsize=(10, 3), dpi=200, facecolor='w', edgecolor='k')
plt.subplot(1, 2, 1)
plt.plot(Dino[:,0], Dino[:,1],'g', lw=1)
plt.title("The original dino");

plt.subplot(1, 2, 2)
plt.plot(Dino[:,0], Dino[:,1],'g', lw=1)
plt.plot(DiNoise[:,0], DiNoise[:,1],'r', lw=1)
plt.title("The noisy dino");
plt.savefig("Dino1.jpg")
```

```python
# Construct the transformation matrices:
I = np.identity(len(Dino), dtype = None).astype(np.int)
L = -2*np.identity(len(Dino), dtype = None).astype(np.int)

# 'Circularly in the first and the last row': start from first element in the matrix
for x in range(0,len(Dino)):
    L[x-1,x] = 1
    L[x,x-1] = 1
    
DinoSmooth = (np.linalg.inv(I - 0.5*L) )@DiNoise
DinoSmooth[-1,:] = DinoSmooth[0,:] # Close the curve


plt.figure(num=3, figsize=(10, 3), dpi=200, facecolor='w', edgecolor='k')
plt.subplot(1, 2, 1)
plt.plot(Dino[:,0], Dino[:,1],'g', lw=1)
plt.plot(DinoSmooth[:,0], DinoSmooth[:,1],'b',lw=1)
plt.title("The smooth dino for $\lambda$ = 0.5");

DinoSmooth = (np.linalg.inv(I - 5*L) )@DiNoise
DinoSmooth[-1,:] = DinoSmooth[0,:]

plt.subplot(1, 2, 2)
plt.plot(Dino[:,0], Dino[:,1],'g', lw=1)
plt.plot(DinoSmooth[:,0], DinoSmooth[:,1],'b',lw=1)
plt.title("The smooth dino for $\lambda$ = 5");
plt.savefig("Dino2.jpg")


# Check that the transformation with \lambda = 0.5 converges points to the average of their neighbours:
print((DinoSmooth[0,] + DinoSmooth[2,])/2)
print(DinoSmooth[1,])

# We can also see that the curve starts shrinking for larger \lambda
```

```python
Lb = -6*np.identity(len(Dino), dtype = None).astype(np.int)

# 'Circularly in the first and the last row': start from first element in the matrix
for x in range(0,len(Dino)):
    Lb[x-1,x] = 4
    Lb[x,x-1] = 4
    Lb[x-2,x] = -1
    Lb[x,x-2] = -1

DinoSmoother = (np.linalg.inv(I - 0.5*L - 0.9*Lb) )@DiNoise
DinoSmoother[-1,:] = DinoSmoother[0,:]

plt.figure(num=4, figsize=(10, 3), dpi=200, facecolor='w', edgecolor='k')
plt.subplot(1, 2, 1)
plt.plot(Dino[:,0], Dino[:,1],'g', lw=1)
plt.plot(DinoSmoother[:,0], DinoSmoother[:,1],'b',lw=1)
plt.title("The smoother dino for $\\alpha$ = 0.5 & $\\beta$ = 0.9");

plt.subplot(1, 2, 2)
DinoSmoother = (np.linalg.inv(I - 2*L - 2*Lb) )@DiNoise
DinoSmoother[-1,:] = DinoSmoother[0,:]
plt.plot(Dino[:,0], Dino[:,1],'g', lw=1)
plt.plot(DinoSmoother[:,0], DinoSmoother[:,1],'b',lw=1)
plt.title("The even smoother dino for $\\alpha$ = $\\beta$ = 2");
plt.savefig("Dino3.jpg")

```

### 1.1.7. Bacterial Growth





```python
from skimage.color import rgb2gray

# Wie can export the movie frames in Photoshop or similar software to individual files. Analysis is then performed on 
# consecutive images. We will export 54 frames (@ 10 fps).

Areas = []
for index in range(53): 
    Bacteria = skimage.io.imread("ExportMovie/listeria_movie0" + "%02d" %index + ".jpg");
    Bacteria = rgb2gray(Bacteria)
    Bacteria = skimage.filters.prewitt(Bacteria, mask=None) # Prewit or Sobel  
    Bacteria = skimage.filters.gaussian(Bacteria, sigma=1, output=None, preserve_range=True)

    T = skimage.filters.threshold_otsu(Bacteria) # Best thresholding results in this setup.
    Binary = Bacteria > T # Perform thresholding.
    [w,h] = Binary.shape
    Areas = np.append(Areas, np.sum(np.sum(Binary))/(w*h)) # Compute area as %of image resolution.

```

```python
# Plot in different code section
plt.figure(num=None, figsize=(7, 3), dpi=200, facecolor='w', edgecolor='k')
plt.plot(np.linspace(0, 5.4, num=53), Areas, 'r',lw=2);
plt.grid(b=None, which='major', axis='both')
plt.xlabel("Time [s]");
plt.ylabel("Area");
plt.savefig("Area.jpg")

```

## 1. Scale Space

First, we define functions that generate 1D gaussian and laplacian kernels of size t and a helper function that convolves Images with a 1D filter kernel.  

The GaussFilter creates a gaussian kernel $G(x)$ given by  

$$ 
G(x)=\frac{1}{\sqrt{2 \pi t}} e^{\frac{-x^{2}}{2 t}}
 $$  
 
The degree of smoothing is determined by the variance $t = \sigma^2$. A kernel size of $5t$ is chosen.  

The Laplacian of a Gaussian is given by it's second derivative:
$$ 
\operatorname{LoG}(x)= \frac{\mathrm{d}^{2} G}{\mathrm{d} x^{2}}
 $$  
 
which evaluates to:

$$ 
\operatorname{LoG}(x)=-\frac{1}{\pi t^{2}}\left[1-\frac{x^{2}}{2 t}\right] e^{-\frac{x^{2}}{2 t}}
 $$

```python
import numpy as np
import math

# Construct function that outputs kernel of g(x) = 1/(sqrt{2*pi*t}) * exp{ (-x^2)/(2*t)}.

def GaussFilter(t):
    # Declare correct kernel size: 5 times variance. Allocate space for K and x:
    Ksize = 1 + 2*math.ceil(5*math.sqrt(t))
    # Allocate space
    center = math.ceil(Ksize/2)
    x = np.arange(1,Ksize+1) - center
    K = 1/(math.sqrt(2*math.pi*t)) * np.exp( (-x**2)/(2*t) )
    K = np.reshape(K, (1,K.size))
    return K/np.sum(K)

def LaplaceFilter(t):
    Ksize = 1 + 2*math.ceil(5*math.sqrt(t))
    center = math.ceil(Ksize/2)
    x = np.arange(1,Ksize+1) - center
    K = -1/(np.sqrt(2*np.pi*t**3)) * (1 - x**2)/t * np.exp( (1-x**2)/(2*t) )
    K = np.reshape(K, (1,K.size))
    return K/np.sum(K)
```

A helper function `Convol` convolves a selected image with a chosen filter in both image directions.  
It can be shown that according to the convolution theorem, a successive convolution of Gaussians with variances $t_1$ and $t_2$ is equal to a single convolution with a variance of $(t_1+t_2)$.  

The cell below produces a visual summary of these basics.

```python
import skimage
import scipy.ndimage as snd
from matplotlib import pyplot as plt
%matplotlib inline

def Convol(im, filt):
    # Make function to perform 1D convolution twice:
    im = snd.convolve(im.astype(float), filt)
    im = snd.convolve(im.astype(float), filt.T)
    return im

t1 = 40
t2 = 22

fig, ax = plt.subplots(figsize=(15,4), ncols=3)
G1 = GaussFilter(t1)
ax[0].imshow(G1)
ax[0].set(title=('Gaussian with t = ' + str(t1)))
ax[0].axis('off')

L2 = LaplaceFilter(t1)
ax[1].imshow(L2)
ax[1].set(title=('Laplacian with t = ' + str(t2)))
ax[1].axis('off')

ax[2].plot(np.arange(-40,41), GaussFilter(t1+t2).T, label='g(62)');
ax[2].scatter(np.arange(-32,33), snd.convolve(GaussFilter(t1),GaussFilter(t2)).T, 
              s = 10, c = 'r', label='g(40)*g(22)');
ax[2].set(title=('Convolutions of Gaussians, g(sigma)'));
ax[2].grid(b=None, which='major', axis='both', alpha = 0.2)
ax[2].legend();
```

## 2. Detecting blobs at one scale

```python
# Construct a function that finds the extrema in a given image.

def FindExtrema(Im):
    # Load image and allocate space and parameters:
    Im = skimage.color.rgb2gray(Im)
    h,w = Im.shape
    Extrema = []
    Maxima = np.zeros((h,w)).astype(np.float)
    Minima = np.zeros((h,w)).astype(np.float)
    
    # Slice the image for every neigbouring pixel and perform a comparison:
    for i in range(-3,4):
            for j in range(-3,4):
                TempMax = Im[3:(h-3), 3:(w-3)] > Im[(3+i):(h-3+i), (3+j):(w-3+j)]
                TempMin = Im[3:(h-3), 3:(w-3)] < Im[(3+i):(h-3+i), (3+j):(w-3+j)]
                
                # If larger than one neighbour, add 1 to counter:
                Maxima[3:(w-3),3:(h-3)] += TempMax.astype(float)
                Minima[3:(h-3),3:(w-3)] += TempMin.astype(float)
    
    # Find pixels whose counter count to eight:

    Extrema = np.nonzero((np.logical_or(Maxima.T==48,Minima.T==48)))
    # remove misclassified

    return np.asarray(Extrema)

```

```python
# Filter an image at different scales and find the image extrema that have 
# the largest pixel values across the different scales. Make this method a 
# function to generalize in the future.

def FindScale(im, sStart, sEnd, t):
    # im = image, sStart = scale where filtering starts, sEnd = final scale, t = scale step size.
    # Output = a 3xn vector containing location of the extrema in the image and the scale (= blob radius):
    Blobs = np.zeros((3,0))
    
    # Expected number of scales:
    steps = np.round((sEnd-sStart)/t).astype(int)

    # Start with filtering the image at initial scale sStart and finding the extrema. 
    # Save the Gaussian filtered image to perform consecutive filtering with scale t:
    im_G = Convol(im, GaussFilter(sStart))
    im_L = Convol(im_G, LaplaceFilter(sStart))
    E = FindExtrema(im_L)

    # Save maximum values in a (h,w,3) array. We will use this array to compare pixel values at the extrema: 
    h,w = im.shape
    PixVals = np.zeros((h,w,3))
    PixVals[E[0,:], E[1,:], 2] = im_L[E[0,:], E[1,:]]
    
    # Now perform consecutive filtering with the same kernels (define only once): 
    G = GaussFilter(t)
    L = LaplaceFilter(t)

    for i in range(0,steps+1):
        # Apply until steps + 1 to include scale sEnd.
        scale = sStart + (i+1)*t

        # Allocate new space in PixVals:
        PixVals[:, :, 0] = PixVals[:, :, 1]
        PixVals[:, :, 1] = PixVals[:, :, 2]
        
        im_G = Convol(im_G, G)
        im_L = Convol(im_G, L)
    
        E = FindExtrema(im_L)
        PixVals[E[0,:], E[1,:], 2] = im_L[E[0,:], E[1,:]]
        
        # Now look at the previous scale and compare sizes:
        Logic = np.logical_and(PixVals[:, :, 1] > PixVals[:, :, 0], PixVals[:, :, 1] > PixVals[:, :, 2])
        Temp = np.asarray(np.where(Logic == True))
        Temp = np.vstack([Temp, (scale-t)*np.ones( (1, len(Temp[1,:])) )])
        #temp = np.reshape(temp, (2,len(temp)))
        Blobs = np.hstack([Blobs, Temp])
    return np.asarray(Blobs)
    
# Test the scale finder function:
import matplotlib
im = skimage.io.imread("test_blob_uniform.png").astype(np.int);


FoundBlobs = FindScale(im, 20, 80,10)
plt.imshow(im, cmap='gray');
plt.scatter(FoundBlobs[0,:], FoundBlobs[1,:],marker='o', s=500, 
            facecolor='none', edgecolor='r', linewidths=2);
plt.title('Detected blobs marked with red circle');
```

```python
from math import sqrt
from skimage import io
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray


image = io.imread("test_blob_uniform.png");
image_gray = rgb2gray(image)

blobs_log = blob_log(image_gray, max_sigma=30, num_sigma=10, threshold=.1)

# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

blobs_dog = blob_dog(image_gray, max_sigma=30, threshold=.1)
blobs_dog[:, 2] = blobs_dog[:, 2] * sqrt(2)

blobs_doh = blob_doh(image_gray, max_sigma=30, threshold=.01)

blobs_list = [blobs_log, blobs_dog, blobs_doh]
colors = ['blue', 'lime', 'red']
titles = ['Laplacian of Gaussian', 'Difference of Gaussian',
          'Determinant of Hessian']
sequence = zip(blobs_list, colors, titles)

fig, axes = plt.subplots(1, 3, figsize=(9, 3), sharex=True, sharey=True)
ax = axes.ravel()

for idx, (blobs, color, title) in enumerate(sequence):
    ax[idx].set_title(title)
    ax[idx].imshow(image_gray, interpolation='nearest', cmap='gray')
    for blob in blobs:
        y, x, r = blob
        c = plt.Circle((x, y), r, color=color, linewidth=2, fill=False)
        ax[idx].add_patch(c)
    ax[idx].set_axis_off()

plt.tight_layout()
plt.show()
```

## SIFT Analysis

```python
def rotate(origin, point, angle):
    """
    Rotate a point counterclockwise by a given angle around a given origin.

    The angle should be given in radians.
    """
    px, py = point
    ox, oy = origin

    qx = ox + math.cos(angle) * (px - ox) - math.sin(angle) * (py - oy) 
    qy = oy + math.sin(angle) * (px - ox) + math.cos(angle) * (py - oy)
    return qx, qy
```

```python
# Generate a toy dataset to perform SIFT analysis

mean=[0,0]
cov=[[1,0],[0,1]]
N = 50

x,y = np.random.multivariate_normal(mean,cov,N).T
xy = np.vstack((x,y))

# scale and shift all points
xscale = 0.5
yscale = 0.8
xshift = 1
yshift = -2
# index ss for "Scaled and Shifted"
xss= x * xscale + xshift
yss= y * yscale + yshift

xyss = np.vstack((xss,yss))

# rotate the array
theta = 20   # degrees counterclockwise
theta_rad = theta * np.pi/180

xyrot = rotate([0,0], xyss, theta_rad)


# Plots
figures = 3
axis = (-5,5);

f, ax = plt.subplots(ncols=figures, figsize=(16,5))
ax[0].plot(x,y, 'x');
ax[0].scatter(np.mean(x), np.mean(y), color='blue', label='Sample mean')
ax[0].set(title='Random Scatter')

ax[1].plot(x,y, 'x', color='k', alpha=0.5);
ax[1].plot(xss, yss, 'o', color='r', alpha=0.4);
ax[1].scatter(np.mean(xss), np.mean(yss), color='k', label='Shifted mean')
ax[1].set(title='Points shifted by [x+1, y-2] \nand scaled by [x*0.5, y*0.8]')
ax[1].plot((x, xss), (y, yss), color='green', linewidth=0.5);

ax[2].plot(xyrot[0], xyrot[1],'o',color='r',alpha=0.5);
ax[2].set(title='Points rotated by theta = '+ str(round(theta, 1)) + ' degrees')
ax[2].plot(x,y, 'x', color='k', alpha=0.6);
ax[2].plot((x, xyrot[0]), (y, xyrot[1]), color='green', linewidth=0.5);

for figure in range(figures):
    ax[figure].axvline(0, color='k', alpha=0.5, linewidth=0.4)
    ax[figure].axhline(0, color='k', alpha=0.5, linewidth=0.4)
    ax[figure].set(xlim=axis, ylim=axis);
```

The rotation is calculated by analyzing the principal components of the original and transformed data points.

The covariance matrix C is defined as:

$$C = \sum_{i=1}^{n} (p_i-\mu_p)(q_i-\mu_q)^T$$

```python
C = np.cov(xyrot)
C
```

```python
# Decompose C by SVD
U, S, VT = np.linalg.svd(C)
```

Compute 2x2 rotation matrix   
$R = UV^T$  

\begin{array}{@{}cc}
\cos a & -\sin a \\
\sin a & \cos a \\
\end{array}

```python
R = U*VT
R
```
